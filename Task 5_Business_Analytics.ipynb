### State wise Sales, Profit and Discount analysis 

```python
df1 = df.groupby(['State'])[['Sales', 'Discount','Profit']].mean()
df1
```

```python
df2 =  df['State'].value_counts().to_frame()
df1 = pd.merge(df1, df2, left_index=True, right_index=True)
df1
```

```python
df1.rename(columns={'State':'NoOfDeals'},inplace = True)

plt.figure(figsize = (20,14))
sns.violinplot(data = df, x='State', y= 'Sales',palette='muted')
#plt.ylim(-1000,2300)
plt.xticks(rotation = 90)
plt.ylabel('Average Sales in Dollars')
plt.xlabel('States')
plt.show()
```

```python
df1.sort_values('Profit')[['NoOfDeals','Sales','Profit']].plot(kind='bar', figsize= (20,10),label=['NoOfDeals','Avg Sales Price($)','Profit($)'])
plt.ylabel('No Of Deals/Average Sales Price($) and Profit per State($)')
plt.xlabel('States')
plt.legend()
plt.show()
```

```python
# Sorted on basis of Discount
df1.sort_values('Profit')[['Discount','Profit']].plot(kind='bar', figsize= (20,10))
plt.ylabel('Discount($) and Profit($) per State')
plt.xlabel('States')
plt.legend()
plt.show()
```

### We can see we can get maximum profit in Vermont State and Minimum in Ohio.

In Vermont No of Deals is very less still Sales in Dollars and Profit is much more. This indicates the quality of the deals is too good. (Maybe the expensive goods are bought more at Vermont)
In Ohio Profit is negative and loss is maximum. Issues faced in Ohio must be addressed and solve to perform better. 

### City wise Sales analysis

```python
df1 = df.groupby(['City'])[['Sales', 'Discount','Profit']].mean()
df1
```

```python
df2 =  df['City'].value_counts().to_frame()
df1 = pd.merge(df1, df2, left_index=True, right_index=True)
df1
```

```python
df1.rename(columns={'City':'NoOfDeals'},inplace = True)
```

```python
# Sorted on basis of Profit in Dollars. Plotted the 50 Cities with Best Sales
df1.sort_values('Profit').tail(50)[['NoOfDeals','Sales','Profit']].plot(kind='bar',figsize= (20,10),label=['NoOfDeals','Avg Sales Price($)','Profit($)'])
plt.ylabel('No Of Deals/Average Sales Price($) and Profit per City($)')
plt.xlabel('Cities')
plt.legend()
plt.show()
```

```python
# Sorted on basis of Profit in Dollars. Plotted the 50 Cities with least Sales
df1.sort_values('Profit').head(50)[['NoOfDeals','Sales','Profit']].plot(kind='bar',figsize= (20,10),label=['NoOfDeals','Avg Sales Price($)','Profit($)'])
plt.ylabel('No Of Deals/Average Sales Price($) and Profit per City($)')
plt.xlabel('Cities')
plt.legend()
plt.show()
```

```python
# Sorted on basis of Profit.Plotted the 50 Cities with Best Profit
df1.sort_values('Profit')[['Discount','Profit']].tail(50).plot(kind='bar', figsize= (20,10))
plt.ylabel('Discount($) and Profit($) per City')
plt.xlabel('Cities')
plt.legend()
plt.show()
```

```python
# Sorted on basis of Profit.Plotted the 50 Cities with Least Profit
df1.sort_values('Profit')[['Discount','Profit']].head(50).plot(kind='bar', figsize= (20,10))
plt.ylabel('Discount($) and Profit($) per City')
plt.xlabel('Cities')
plt.legend()
plt.show()
```

### We can see we can get maximum profit in Jamestown City and minimum in Bethlehem.

In Jamestown No of Deals is very less still Sales in Dollars and Profit is much more. This indicates the quality of the deals is too good. (Maybe the expensive goods are bought more at Vermont)
In Bethlehem Profit is negative and loss is maximum
